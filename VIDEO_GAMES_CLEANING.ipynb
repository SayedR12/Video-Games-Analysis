**PREPROCESSING**

```python
#fill release wite the most frequent value
df['release_date'].fillna(df['release_date'].mode()[0],inplace=True)
```

```python
df['release_date']
```

```python
df['release_date']=pd.to_datetime(df['release_date'])
```

```python
df.dtypes
```

```python
#fill critic_score with average
df['critic_score'].fillna(df['critic_score'].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
#fill withe date from relase date before that i should convert datatype for last_ubdate
df['last_ubdate']=pd.to_datetime(df['last_update'])
df['last_update'].fillna(df['release_date'], inplace=True)
```

```python
# i will asume no sales and i will fill with 0
df[['na_sales', 'jp_sales', 'pal_sales', 'other_sales']] = df[['na_sales', 'jp_sales', 'pal_sales', 'other_sales']].fillna(0)
```

```python
df.isnull().sum()
```

```python
#calculate total sales after fill nulls
df['total_sales'] = df[['na_sales', 'jp_sales', 'pal_sales', 'other_sales']].sum(axis=1)
```

```python
#fill developer with publisher bassed on dataset and publisher and developer have same name
df['developer'] = df['developer'].fillna(df['publisher'])
```
